```python
# time_course_df = pd.read_excel('Time-course_2.xlsx', 'sheet_name = 0 is default', index_col=None, na_values=[0], skiprows=6) # Read excel 'Time-course_2.xlsx' with sheet 'S1(1-342)-C  + Y.hCAD' to dataframe
time_course_df = pd.read_excel('Time-course_2.xlsx', sheet_name=0, index_col=0, parse_dates=True, skiprows=6) # set indexed column 'Timestamp' and dates in excel is parsed as datatime type
```

```python
# CFP data frame till # 75
```

```python
cfp_df = time_course_df.iloc[0:75]
```

```python
cfp_df.head() # print first 5 rows customize using .head(n) n is number of rows you want to see
```

```python
# Some basic statics of the CFP data
cfp_df.describe()
```

```python
# YFP part starts here # 156
```

```python
yfp_df = time_course_df.iloc[156:231]
yfp_df.tail()
```

```python
# E-FRET part starts here # 234
```

```python
efret_df = time_course_df.iloc[234:]
efret_df.tail()
```

```python
# cleaning the empty value -- to 0.00
efret_df = efret_df.replace('--', 0)
```

```python
# Finding the Background column containing '-1.#IND0'
background_col = [col for col, ser in efret_df.iteritems() if ser.astype(str).str.match('-1.#IND0*').any()] # find the background column from the E-FRET data frame
background_col
```

```python
# background column's index
background_col_index = efret_df.columns.get_loc(background_col[0])
background_col_index
```

```python
# Corrected CFP data frame
corrected_cfp_df = pd.DataFrame()
for c in cfp_df.columns:
    corrected_cfp_df[c] = cfp_df.apply(lambda row: row[c] - row[cfp_df.columns[background_col_index]], axis=1)
corrected_cfp_df.head()
```

```python
# Corrected YFP data frame
corrected_yfp_df = pd.DataFrame()
for c in cfp_df.columns:
    corrected_yfp_df[c] = yfp_df.apply(lambda row: row[c] - row[yfp_df.columns[background_col_index]], axis=1)
corrected_yfp_df.head()
```

```python
# Ratio of corrected_yfp_df / corrected_cfp_df
ratio_df = pd.DataFrame()
ratio_df = corrected_yfp_df.div(corrected_cfp_df, level=1, fill_value=0)
ratio_df.head()
```

```python
ratio_df = ratio_df.iloc[[0, -1]]
ratio_df
```

```python
# Delete column at index background_col_index
ratio_df_clean = ratio_df.drop([ratio_df.columns[background_col_index]], axis='columns')
ratio_df_clean
```

```python
ratio_df_clean.columns = ratio_df_clean.columns.str.replace(r" Mean Intensity \(.*\), ch1:CFP","")
ratio_df_clean
```

```python
# Remove index of 'TimePoint'
ratio_df_clean.index.names = [None]
final_ratio_df_clean = ratio_df_clean.T
final_ratio_df_clean
```

```python
# Rename column to 'initial' and 'final'
final_ratio_df_clean = final_ratio_df_clean.rename(lambda x: 'YFP/CFP(Initial)' if x == 0 else 'YFP/CFP(Final)', axis=1)
final_ratio_df_clean
```

```python
# E-FRET cleaning here initial and final only
```

```python
efret_df = efret_df.iloc[[0, -1]]
efret_df
```

```python
# Delete column at index background_col_index
efret_df_clean = efret_df.drop([efret_df.columns[background_col_index]], axis='columns')
efret_df_clean
```

```python
efret_df_clean.columns = efret_df_clean.columns.str.replace(r" Mean Intensity \(.*\), ch1:CFP","")
efret_df_clean
```

```python
# Remove index of 'TimePoint'
efret_df_clean.index.names = [None]
final_efret_df_clean = efret_df_clean.T
final_efret_df_clean
```

```python
# Rename column to 'initial' and 'final'
final_efret_df_clean = final_efret_df_clean.rename(lambda x: 'E-FRET(Initial)' if x == 0 else 'E-FRET(Final)', axis=1)
final_efret_df_clean
```

```python
# Joining final_ratio_df_clean and final_efret_df_clean as their Columns are unique it will not show any issue
```

```python
combined_df = final_ratio_df_clean.join(final_efret_df_clean, lsuffix="_YFP/CFP", rsuffix="_E-FRET")
combined_df
```
